```python
#30/05/25
#linked list home work
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
    def insert_at_beginning(self, data):
        print(f"Inserting {data} at the beginning...")
        new_node = Node(data)          
        new_node.next = self.head      
        self.head = new_node           
    def insert_at_end(self, data):
        print(f"Inserting {data} at the end...")
        new_node = Node(data)
        if self.head is None:
            self.head = new_node       
            return
        temp = self.head
        while temp.next:
            temp = temp.next          
        temp.next = new_node           
    def insert_at_position(self, position, data):
        print(f"Inserting {data} at position {position}...")
        if position == 0:
            self.insert_at_beginning(data)
            return
        new_node = Node(data)
        temp = self.head
        for _ in range(position - 1):
            if temp is None:
                print("Position out of range")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node
ll = LinkedList()
ll.insert_at_end(10)
ll.insert_at_end(20)
ll.display()
ll.insert_at_beginning(5)
ll.display()
ll.insert_at_position(2, 15)
ll.display()
```

```python
#deletion in linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
    def delete_from_beginning(self):
        if self.head is None:
            print("List is empty")
            return
        print(f"Deleting {self.head.data} from beginning...")
        self.head = self.head.next
    def delete_from_end(self):
        if self.head is None:
            print("List is empty")
            return
        if self.head.next is None:
            print(f"Deleting {self.head.data} (only node)...")
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        print(f"Deleting {temp.next.data} from end...")
        temp.next = None  # Remove last node

    # Step 3: Delete at position
    def delete_at_position(self, position):
        if self.head is None:
            print("List is empty")
            return
        if position == 0:
            print(f"Deleting {self.head.data} from position 0...")
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if temp.next is None:
                print("Position out of range")
                return
            temp = temp.next
        if temp.next is None:
            print("Position out of range")
            return
        print(f"Deleting {temp.next.data} from position {position}...")
        temp.next = temp.next.next
ll = LinkedList()
ll.head = Node(10)
ll.head.next = Node(20)
ll.head.next.next = Node(30)
ll.head.next.next.next = Node(40)
ll.display()
ll.delete_from_beginning()
ll.display()
ll.delete_from_end()
ll.display()
ll.delete_at_position(1)
ll.display()
```
